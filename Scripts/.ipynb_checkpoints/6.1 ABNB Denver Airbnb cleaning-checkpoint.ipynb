# Data Cleaning

```python
#creating a subset with only the columns I wish to keep 

columns_to_keep = ['id','host_id','host_since', 'host_location', 'host_response_time', 'host_response_rate', 'host_acceptance_rate', 'host_is_superhost','neighbourhood_cleansed', 'latitude', 'longitude', 'property_type', 'room_type', 'accommodates', 'bathrooms_text', 'beds', 'price', 'minimum_nights', 'maximum_nights','availability_30', 'availability_60', 'availability_90', 'availability_365', 'number_of_reviews','number_of_reviews_ltm', 'number_of_reviews_l30d', 'first_review', 'last_review', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'instant_bookable', 'reviews_per_month']
df_bnb_selected = df_airbnb[columns_to_keep]
```

# Missing Data 

```python
#finding missing values 
df_bnb_selected.isnull().sum()
```

```python
#going to drop the 63 rows in beds since this isn't a large portion of the data set 
df_bnb_selected.dropna(subset = ['beds'], inplace = True)
```

```python
#all of the review categories represent a lack of reviews. For now I am going to impute a value of 0 to represent has received no reviews. I looked at the data in excel and confirmed that these columns currently do not have a value of 0 anywhere so 0 can truly reprsent a lack of reviews. This will skew the average so the mode will be more informative when looking at descriptive statistics but moving forward I can also create a subset removing the rows where the column value = 0. I think I may make a flagged column for has reviews and doesn't have reviews in the future so I don't want to impute the average and I don't want to get rid of these columns completely. As I get further in my analysis I will decide how to handle these columns.
```

```python
#creating a variable for all of the columns I want to impute 0 for the null value 
columns_to_impute = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 
                     'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
                     'review_scores_value', 'reviews_per_month']
```

```python
#Imputing 0 for the null value
```

```python
df_bnb_selected[columns_to_impute] = df_bnb_selected[columns_to_impute].fillna(0)
```
